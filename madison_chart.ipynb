```python
# Madison's Markdown, awesome!
# Scatter
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df = pd.read_excel ('faa_data_model.xlsx', sheet_name= 'clean_data', index_col='Combination')
```

```python
df_WBS_mean = pd.DataFrame(df.groupby('WBS').mean()['Actual Expended Sum'])
```

```python
df_WBS_mean.plot(kind= 'hist', bins = 20, density = False, facecolor = 'g')
plt.xlabel ('Expended Sum - Actual')
plt.title('Mean Actual Expended Sum by WBS')
plt.ylabel('Freq')
plt.show()
```
